```python
import pandas as pd
import datetime
import numpy as np
from sqlalchemy import create_engine
# from sqlalchemy import 
```

```python
url = "https://raw.githubusercontent.com/ahyds/ETL_project/master/Resources/Van-data%202014-2018.csv"
van_data = pd.read_csv(url)
```

```python
van_data.head()
```

```python
new_van_data = van_data[['TYPE', 'YEAR', 'MONTH','HOUR']].copy()
new_van_data.head(10)
```

```python
new_van_data['HOUR'].replace({NaN: 100})
new_van_data.head(10)
```

```python
new_van_data['DAY/NIGHT'] = np.where((new_van_data['HOUR']>=18  | (np.where(new_van_data['HOUR']<6 )) ,'NIGHT',(np.where((new_van_data['HOUR']>=6)&(new_van_data['HOUR']<18)),'DAY','NaN')))
new_van_data.head(10)
```

```python
new_van_data['DAY/NIGHT'] = np.where((new_van_data['HOUR']<18) & (new_van_data['HOUR']>=6),'DAY',np.where(new_van_data['HOUR'] == 'NaN','NaN','NIGHT'))
new_van_data.head(10)
```

```python
new_van_data.describe()
```

```python
year_count = new_van_data.groupby ('YEAR').count()
year_count
```

```python
month_mean = new_van_data.groupby(['MONTH']).count()
month_mean
```

```python
day_night = new_van_data.groupby(['DAY/NIGHT']).count()
day_night
```

```python
type_count = new_van_data['TYPE'].value_counts()
type_count
```

```python
new_van_data['TYPE'] = new_van_data['TYPE'].replace({'Theft from Vehicle':'Other Theft','Mischief':'Assault','Break and Enter Residential/Other':'Break and Enter','Theft of Bicycle':'Other Theft',
                                                     'Break and Enter Commercial':'Break and Enter','Vehicle Collision or Pedestrian Struck (with Injury)':'Traffic','Theft of Vehicle':'Auto Theft','Vehicle Collision or Pedestrian Struck (with Fatality)':'Traffic'})
```
